Q1: **Define the following with an example:**  
1) **Artificial Intelligence**  
2) **Machine Learning**  
3) **Deep Learning**

A1:  
- **Artificial Intelligence (AI):** AI refers to the simulation of human intelligence processes by machines, especially computer systems. One example is IBM's Watson, which is capable of understanding natural language and answering questions posed in human language.

- **Machine Learning (ML):** ML is a subset of AI that enables machines to learn from data without being explicitly programmed. An example is a spam email filter that learns to distinguish between spam and non-spam emails based on examples it's trained on.

- **Deep Learning (DL):** DL is a subset of ML that uses artificial neural networks with many layers (deep neural networks) to model and understand complex patterns in data. An example is Google's AlphaGo, which uses deep learning to play the board game Go at a superhuman level.

Q2: **What is supervised learning? List some examples of supervised learning.**

A2:  
- **Supervised Learning:** Supervised learning is a type of machine learning where the model is trained on labeled data, meaning each input data point is associated with the correct output. The model learns to map inputs to outputs based on examples. Examples include:
  - Image classification: Given images of animals labeled as "cat" or "dog," the model learns to classify new images.
  - Spam detection: Training a model on labeled emails (spam or not spam) to classify new incoming emails.
  - Predicting house prices: Using features like square footage, number of bedrooms, etc., to predict the price of a house.

Q3: **What is unsupervised learning? List some examples of unsupervised learning.**

A3:  
- **Unsupervised Learning:** Unsupervised learning is a type of machine learning where the model is trained on unlabeled data, and it must find patterns and structures within the data on its own. Examples include:
  - Clustering: Grouping similar data points together without any prior knowledge of which group they belong to.
  - Anomaly detection: Identifying unusual patterns or outliers in data that do not conform to expected behavior.
  - Dimensionality reduction: Reducing the number of features in a dataset while preserving its important structure and relationships.

Q4: **What is the difference between AI, ML, DL, and DS?**

A4:  
- **Artificial Intelligence (AI)** is the broader concept of machines being able to carry out tasks in a way that we would consider "smart."
- **Machine Learning (ML)** is a subset of AI that focuses on the development of algorithms that allow computers to learn from and make predictions or decisions based on data.
- **Deep Learning (DL)** is a subset of ML that utilizes neural networks with many layers (deep neural networks) to model and understand complex patterns in large amounts of data.
- **Data Science (DS)** is an interdisciplinary field that uses scientific methods, algorithms, processes, and systems to extract insights and knowledge from structured and unstructured data.

Q5: **What are the main differences between supervised, unsupervised, and semi-supervised learning?**

A5:  
- **Supervised Learning:** Requires labeled data and involves training a model to make predictions based on that data.
- **Unsupervised Learning:** Involves training a model on unlabeled data to find hidden patterns or intrinsic structures.
- **Semi-Supervised Learning:** Utilizes both labeled and unlabeled data for training, which can be useful when labeled data is scarce or expensive.

Q6: **What is train-test-validation split? Explain the importance of each term.**

A6:  
- **Train-Test-Validation Split:** In machine learning, the dataset is typically divided into three subsets:
  - **Training set:** Used to train the model.
  - **Test set:** Used to evaluate the model's performance and assess its generalization to new, unseen data.
  - **Validation set:** Used to fine-tune the model's hyperparameters and prevent overfitting.

The importance of each term:
- **Training set:** Used to train the model, allowing it to learn patterns and relationships within the data.
- **Test set:** Used to evaluate the model's performance on unseen data and estimate how well it will generalize to new data.
- **Validation set:** Used to fine-tune the model's hyperparameters (such as learning rate, regularization parameter) and prevent overfitting by providing an unbiased evaluation during model development.

Q7: **How can unsupervised learning be used in anomaly detection?**

A7:  
Unsupervised learning can be used in anomaly detection by training a model on normal, unlabeled data and then identifying instances that deviate significantly from the learned patterns as anomalies. Techniques such as clustering, density estimation, and dimensionality reduction can be applied for anomaly detection without the need for labeled anomalous data.
